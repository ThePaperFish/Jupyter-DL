```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
def count(kw):
	pytrends.build_payload([kw], cat=0, timeframe='today 3-m', geo='MY', gprop='')
	list = pytrends.interest_over_time()
	if len(list)> 0:
		return list[kw].sum()
	return 0

def load_txt(path):
	f = open(path, "r")
	content = f.read()
	f.close()
	return content

def nick_name(name):
	index = name.find(')')
	if index == -1:
		return name
	name = name[1 : index]
	name = name.split('/')
	return name[0].strip() 
```

```python
path = r'C:\Users\tlchong\Desktop\all\list\Singerss.txt'
kw_list = load_txt(path).split('\n')
kw_list.sort()

path = r'C:\Users\tlchong\Desktop\all\list\Singers.txt'
ch_list = load_txt(path).split('\n')
ch_list.sort()

result_k = []
result_v = []
```

```python
for i in range(0, len(kw_list)):
	name = nick_name(kw_list[i])
	if name in ch_list:
		continue
    
	c = count(name)
	if c > 0:
		print("{} : {}".format(name,c))
		result_k += [name]
		result_v += [c]

	if i % 100 == 0:
		print('--'* 20 + 'checkpoint {}'.format(i * 100))
```

```python
result_v, result_k = (list(t) for t in zip(*sorted(zip(result_v, result_k))))
result_v.reverse()
result_k.reverse()
print(result_v)
print(result_k)
```
